<h2> Lire les données </h2>
<p>
Nous utilisons la fonction <code> pandas.read_csv () </code> pour lire le fichier csv. Dans le crochet, nous mettons le chemin du fichier avec un guillemet, de sorte que les pandas liront le fichier dans une trame de données à partir de cette adresse. Le chemin du fichier peut être une URL ou votre adresse de fichier local. <br>
Étant donné que les données n'incluent pas les en-têtes, nous pouvons ajouter un argument <code> headers = None </code> dans la méthode <code> read_csv () </code>, afin que les pandas ne définissent pas automatiquement la première ligne comme en-tête . <br>
Vous pouvez également affecter l'ensemble de données à n'importe quelle variable que vous créez.
</p>

```python
# Import pandas library
import pandas as pd

# Lecture du fichier via une URL et son attribution à une variable "df"
#other_path = "https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/DA0101EN/auto.csv"

# Lecture du fichier depuis le disque local et son attribution à une variable "df"
#other_path = r"C:\PYTHON PRACTICE\2_Data Science project\1_Importation des data/auto.csv"
df = pd.read_csv(r"C:\python_pratique\data science\1_Importation des data/auto.csv", header=None)
#df = pd.read_csv(other_path, header=None)
```

<h3> Ajouter des en-têtes </h3>
<p>
Jetez un œil à notre ensemble de données; pandas définit automatiquement l'en-tête par un entier à partir de 0.
</p>
<p>
Pour mieux décrire nos données, nous pouvons introduire un en-tête, ces informations sont disponibles à l'adresse: <a href="https://archive.ics.uci.edu/ml/datasets/Automobile" target="_blank"> https: // archive.ics.uci.edu/ml/datasets/Automobile </a>
</p>
<p>
Ainsi, nous devons ajouter des en-têtes manuellement.
</p>
<p>
Tout d'abord, nous créons une liste "en-têtes" qui inclut tous les noms de colonnes dans l'ordre.
Ensuite, nous utilisons <code> dataframe.columns = headers </code> pour remplacer les en-têtes par la liste que nous avons créée.
</p>

```python
# create headers list
headers = ["symboling","normalized-losses","make","fuel-type","aspiration", "num-of-doors","body-style",
         "drive-wheels","engine-location","wheel-base", "length","width","height","curb-weight","engine-type",
         "num-of-cylinders", "engine-size","fuel-system","bore","stroke","compression-ratio","horsepower",
         "peak-rpm","city-mpg","highway-mpg","price"]
print("headers\n", headers)
```

Nous remplaçons les en-têtes et revérifions notre dataset

```python
df.columns = headers
df.head(10)
```

nous pouvons supprimer les valeurs manquantes dans la colonne "prix" comme suit

```python
df.dropna(subset=["price"], axis=0)
```

Maintenant, nous avons lu avec succès l'ensemble de données brutes et ajouté les en-têtes corrects dans le bloc de données.

<h2> Description des colonnes </h2>
Si nous souhaitons obtenir un résumé statistique de chaque colonne, tel que le nombre, la valeur moyenne de la colonne, l'écart type de la colonne, etc. Nous utilisons la méthode describe:

Cette méthode fournira diverses statistiques récapitulatives, à l'exclusion des valeurs <code> NaN </code> (Not a Number).

```python
df.describe()
```

<p>
Ceci montre le résumé statistique de toutes les colonnes de type numérique (int, float). <br>
Par exemple, l'attribut «symboling» a 205 points, la valeur moyenne de cette colonne est 0,83, l'écart type est 1,25, la valeur minimale est -2, le 25e centile est 0, le 50e centile est 1, le 75e centile est 2 et la valeur maximale est 3.
<br>
Cependant, que se passe-t-il si nous souhaitons également vérifier toutes les colonnes, y compris celles de type objet.
<br> <br>

Vous pouvez ajouter un argument <code> include = "all" </code> à l'intérieur du crochet. Essayons à nouveau.
</p> 

```python
# describe all the columns in "df" 
df.describe(include = "all")
```

<p>
Désormais, il fournit le résumé statistique de toutes les colonnes, y compris les attributs de type objet. <br>
Nous pouvons maintenant voir le nombre de valeurs uniques, qui est la valeur la plus élevée et la fréquence de la valeur la plus élevée dans les colonnes de type objet. <br>
Certaines valeurs du tableau ci-dessus s'affichent sous la forme "NaN", car ces chiffres ne sont pas disponibles pour un type de colonne particulier. <br>
</p>
